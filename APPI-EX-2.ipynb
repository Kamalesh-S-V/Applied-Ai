```python
# Inputs
inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]

# Learning rate
lr = 0.1


# ---------------- OR GATE ----------------
print("OR Gate")

# Initial weights and bias
w1, w2, b = -0.5, 0.7, 0.88

# OR gate target values
targets = [0, 1, 1, 1]

for _ in range(1):  # one epoch
    error = 0
    for i in range(4):
        x1, x2 = inputs[i]

        # Net input
        net = x1 * w1 + x2 * w2 + b

        # Step activation function
        y = 1 if net >= 0 else 0

        # Error calculation
        e = targets[i] - y

        # Weight & bias update
        w1 += lr * e * x1
        w2 += lr * e * x2
        b  += lr * e

        error += abs(e)

    if error == 0:
        break

print("Final OR:", w1, w2, b)


# ---------------- AND GATE ----------------
print("\nAND Gate")

# Reset weights and bias
w1, w2, b = -0.5, 0.7, 0.88

# AND gate target values
targets = [0, 0, 0, 1]

for _ in range(1):  # one epoch
    error = 0
    for i in range(4):
        x1, x2 = inputs[i]

        # Net input
        net = x1 * w1 + x2 * w2 + b

        # Step activation function
        y = 1 if net >= 0 else 0

        # Error calculation
        e = targets[i] - y

        # Weight & bias update
        w1 += lr * e * x1
        w2 += lr * e * x2
        b  += lr * e

        error += abs(e)

    if error == 0:
        break

print("Final AND:", w1, w2, b)
```
